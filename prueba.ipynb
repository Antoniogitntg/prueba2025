Construye una función que reciba una serie de datos (como puede ser una variable en un dataset) y devuelva esa serie normalizada entre 0 y 1. Además la función tiene que 
comprobar que el formato de la variable de entrada es una serie y elevar un error en caso  contrario. Ejecuta esta función sobre la variable que consideres. 

```python
def fun(df):
    if not isinstance(df, pd.Series):
        raise TypeError("La entrada debe ser una Serie de pandas")
    
    normalized_series = (df - df.min()) / (df.max() - df.min())
    return normalized_series


x= pd.Series(src_train['var_1'])
fun(x)


    
```

Genera los test unitarios que consideres para comprobar el funcionamiento correcto de la función anterior. 

```python
dcol_num= src_train.select_dtypes(np.number).columns
dic={}
for col in dcol_num.values:
    x= pd.Series(src_train[col])
    a=fun(x)
    dic[col]=a
    
df = pd.DataFrame(dic)
df
```
