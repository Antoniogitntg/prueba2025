 Realiza un diagrama de cajas múltiple de la variable “var_2” por cada nivel de la variable “var_47”

```python
import matplotlib.pyplot as plt
raw_datos_cle['var_2'] = pd.to_numeric(raw_datos_cle['var_2'], errors='coerce')
raw_datos_cle['var_47'] = raw_datos_cle['var_47'].astype('category')
plt.figure(figsize=(10, 6))
raw_datos_cle.boxplot(column='var_2', by='var_47')
plt.title('Diagrama de cajas de var_2 por niveles de var_47')
plt.suptitle('')
plt.xlabel('var_47')
plt.ylabel('var_2')
plt.show()
```

Si existen outliers en el gráfico, imprímelos por pantalla. 

```python
import numpy as np
from tabulate import tabulate
raw_datos_cle['var_2'] = pd.to_numeric(raw_datos_cle['var_2'], errors='coerce')
raw_datos_cle['var_47'] = raw_datos_cle['var_47'].astype('category')
raw_datos_cle_2_47 = raw_datos_cle[['var_2','var_47']]
raw_datos_cle_2_47_0= raw_datos_cle_2_47[raw_datos_cle_2_47['var_47']=="0"]
q1 = np.percentile(raw_datos_cle_2_47_0['var_2'], 25)
q3 = np.percentile(raw_datos_cle_2_47_0['var_2'], 75)
iqr = q3 - q1
umbra_sup = q3+1.5*iqr
umbra_inf = q1-1.5*iqr
print("PARA LA VARIABLE 47=0\n""q1:",q1,"q3:",q3,"iqr:",iqr,"umbral sup:",umbra_sup,"umbral inf:",umbra_inf,"\n")
outliers = np.where((raw_datos_cle_2_47_0['var_2'] < umbra_inf) | (raw_datos_cle_2_47_0['var_2'] > umbra_sup))
valores=raw_datos_cle_2_47_0['var_2'].iloc[outliers]
print(valores)

raw_datos_cle_2_47_1= raw_datos_cle_2_47[raw_datos_cle_2_47['var_47']=="1"]
q1 = np.percentile(raw_datos_cle_2_47_1['var_2'], 25)
q3 = np.percentile(raw_datos_cle_2_47_1['var_2'], 75)
iqr = q3 - q1
umbra_sup = q3+1.5*iqr
umbra_inf = q1-1.5*iqr
print("PARA LA VARIABLE 47=1\n""q1:",q1,"q3:",q3,"iqr:",iqr,"umbral sup:",umbra_sup,"umbral inf:",umbra_inf)
outliers = np.where((raw_datos_cle_2_47_1['var_2'] < umbra_inf) | (raw_datos_cle_2_47_1['var_2'] > umbra_sup))
valores1=raw_datos_cle_2_47_1['var_2'].iloc[outliers]
print(valores1)
```
