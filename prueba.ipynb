 Localiza las variables con datos nulos (NA, NaN) y decide como tratar estas variables.

```python
numero_nan_x_colum=raw_datos.isna().sum()
numero_nan_x_colum
 
```

```python
 
for col  in numero_nan_x_colum.index:
    if numero_nan_x_colum[col]>0:
        print(f"Columna '{col}' tiene {numero_nan_x_colum[col]} valores NaN.")
```

Elimino la columna "var_70" porque casi todos sus valores son NAN

```python
raw_datos_cle=raw_datos.drop(columns=["var_70"])
```

```python
raw_datos_cle.info()
```

Elimino los NAN de la columna "var_65"

```python
raw_datos_cle.dropna(inplace=True)
```

```python
raw_datos_cle.info()
```

 Recarga los datos como src_train dejando de Python decida el tipo más conveniente, a excepción de la variable ID que debe de ser de tipo string (object).

```python
src_train = raw_datos_cle.convert_dtypes()
src_train.info()
```

```python
src_train['ID'] = src_train['ID'].astype('string')
```

```python
src_train.info()
```

```python
for col in raw_datos_cle.columns:
    src_train[col] = pd.to_numeric(src_train[col], errors='coerce')
```

```python
src_train.info()
```
